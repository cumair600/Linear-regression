# Importing Libraries

```python
import pandas as pd
from matplotlib import pyplot as plt
import numpy as np
```

# Loading Data into dataframe

```python
data = pd.read_csv('data/ex1data1.txt', sep=",", header=None)
data.columns = ["Population", "Profit"]
data
```

# Plotting Scattered data

```python
plt.scatter(data.Population, data.Profit, label= "cross", color= "red",  
            marker= "x", s=30) 
  
plt.ylabel('Profit in $10,000s')  
plt.xlabel('Population of city in $10,000s') 
plt.title('Scattered Plot') 
plt.legend() 
plt.show() 
```

# Gradient Descent

```python
w0 = 1
w1 = 1
m = data.Population.size

for j in range(1000):
    print("Iteration : " + str(j))
    y_i = np.array([])
    # dot product
    for i in range(m):
        dot = w0 + w1 * data.Population[i]
        y_i = np.append(y_i, dot, axis=None)

    # loss calculation
    loss = np.array([])
    for i in range(m):
        l = (y_i[i] - data.Profit[i])**2
        loss = np.append(loss, l)
    loss = loss.sum()    
    loss = 1/(2*m) * (loss)
    print("loss : " + str(loss))

    #update weights
    u = np.array([])
    alpha = 0.001
    for i in range(m):
        l = (y_i[i] - data.Profit[i]) * data.Population[i]
        u = np.append(u, l)
    u = u.sum()
    w0 = w0 - alpha*(1/m)*1
    w1 = w1 - alpha*(1/m)*u
```

```python
plt.plot(data.Population, y_i, label= "prediction", color= "blue")
plt.scatter(data.Population, data.Profit, label= "actual data", color= "red",  
            marker= "x", s=30) 
  
plt.ylabel('Profit in $10,000s')  
plt.xlabel('Population of city in $10,000s') 
plt.title('Trained Model') 
plt.legend() 
plt.show() 
```
